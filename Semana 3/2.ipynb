```python
import pandas as pd 
import numpy as np
import timeit

df = pd.read_csv('../Semana 2/census.csv')
df.head()
```

```python
#como podemos ver obtenemos una sintaxis un poco diferente a la tradicional y mucho mas "pandorable", sin embargo hay
#que recalcar que tambien podriamos hacer lo mismo con una sintaxis tradicional sin ningun problema
df_aux = df[df['SUMLEV'] == 50]
df_aux.set_index(['STNAME', 'CTYNAME'], inplace=True)
df_aux.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
#Ahora veremos un calculo del rendimiento de cada aproximacion usando la libreria timeit
def first_aproach():
    global df
    
    return (df.where(df['SUMLEV'] == 50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
    
df = pd.read_csv('../Semana 2/census.csv')

timeit.timeit(first_aproach, number = 10)
```

```python
#ahora veremos el rendimiento de la segunda aproximancion 
def second_aproach():
    global df
    
    df_aux = df[df['SUMLEV'] == 50]
    df_aux.set_index(['STNAME', 'CTYNAME'], inplace=True)
    return df_aux.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

df = pd.read_csv('../Semana 2/census.csv')

timeit.timeit(second_aproach, number = 10)
```

```python
#como podemos ver el segundo enfoque es mucho mas rapido que el primero, sin embargo es importante conocer ambos dado
#que en ciertos contextos nos podemos encontrar una sintaxis parecida a la primera y asi podremos entender que es lo que se 
#desea hacer
```

FUNCION APPLY


```python
#En algun momento todos hemos la funcion map que nos permite pasar como parametro una funcion y un iterable(listas, etc) y nos
#devuelve como resultado una lista donde se aplico la funcion en la lista objetivo es asi que un acercamiento en los dataframes
#nos permite encontrar la funcion apply que hace algo parecido pero enfocado en dataframes
def min_max(row):
    data = row[['POPESTIMATE2010', 
                'POPESTIMATE2012', 
                'POPESTIMATE2013', 
                'POPESTIMATE2014', 
                'POPESTIMATE2015']]
    
    return pd.Series({'min' : np.min(data), 'max' : np.max(data)})
```

```python
df.apply(min_max, axis='columns').head()
#en este caso podemos observar que aplicamos la min_max que sirve para obtener el maximo y minimo de cada columna dadas las columnas especificadas
#dentro de la variable data
```

```python
#y tambien podemos modificar la funcion de otra manera
#En algun momento todos hemos la funcion map que nos permite pasar como parametro una funcion y un iterable(listas, etc) y nos
#devuelve como resultado una lista donde se aplico la funcion en la lista objetivo es asi que un acercamiento en los dataframes
#nos permite encontrar la funcion apply que hace algo parecido pero enfocado en dataframes
def min_max(row):
    data = row[['POPESTIMATE2010', 
                'POPESTIMATE2012', 
                'POPESTIMATE2013', 
                'POPESTIMATE2014', 
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    
    return row

df.apply(min_max,axis='columns')
```

```python
#tambien podemos tener un ejemplo del uso de apply junto a lambda
row = ['POPESTIMATE2010', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']

df.apply(lambda x : np.max(x[row]), axis=1).head()
```

```python
# La belleza de la función apply radica en que permite flexibilidad para realizar cualquier manipulación deseada,
# ya que la función que pasas a apply puede ser personalizada de la manera que prefieras. Supongamos que queremos
# dividir los estados en cuatro categorías: Noreste, Medio Oeste, Sur y Oeste. Podemos escribir una función
# personalizada que devuelva la región basada en la información de las regiones de los estados, que se obtiene de Wikipedia.

def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
