Mergin dataframes

```python
#interseccion o inner join
import pandas as pd

staff_df = pd.DataFrame([{"Name" : "Kelly", "Role" : "Director of HR"},
                        {"Name" : "Sally", "Role" : "Course Liasons"},
                        {"Name" : "James", "Role": "Grader"}])

staff_df = staff_df.set_index('Name')

student_df = pd.DataFrame([{"Name" : "James", "School" : "Bussines"},
                          {"Name" : "Mike", "School" : "Law"},
                          {"Name" : "Sally", "School" : "Engineering"}])

student_df = student_df.set_index('Name')

print(staff_df)
print(student_df)
```

```python
#ahora q tenemos creados nuestros dataframes si deseamos unirlos necesitamos indicar el indice comun de estos 
#con el siguiente metodo
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
#donde outer es lo que denota el tipo de union de nuestros dataframes en este caso siendo outer igual a la union
```

```python
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
#En este caso inner denota la interseccion que tenemos en nuestros dataframes
```

```python
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
#En este nos devolvera la totalidad del primer dataframe junto a todos sus datos existentes en ambos dataframes
#en caso de que no tenga un dato existente en el segundo dataframe este simplemente se pondr un valor de NaN
```

```python
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
#En este caso se obtendran la totalidad de los datos del segundo dataframe existentes en ambos dataframes
```

```python
#Sin embargo esta no es la unica forma de manejar operaciones de conjunto con dataframes, ni tampoco es necesario
#utilizar siempre un indice , sino simplemente indicando la columna en el parametro on
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()

pd.merge(staff_df, student_df, how = 'right', on='Name')
```

```python
#ahora veremos lo q sucede cuando tenemos conflictos entre dataframes
staff_df = pd.DataFrame([{'Name' : 'Kelly', 'Role' : 'Director of HR', 'Location' : 'State Street'},
                         {'Name' : 'Sally', 'Role' : 'Course of Liason', 'Location' : 'Washington Avenue'},
                         {'Name' : 'James', 'Role' : 'Grader', 'Location' : 'Washington Avenue'}])

student_df = pd.DataFrame([{'Name' : 'James', 'School' : 'Bussines', 'Location' : '1024 Billiard Avenue'},
                           {'Name' : 'Mike' , 'School' : 'Law', 'Location' : 'Fraternity House #22'},
                           {'Name' : 'Sally' , 'School' : 'Enginneering', 'Location' : '512 Wilson Crescent'}])

pd.merge(staff_df, student_df, how ='left', on='Name')
```

```python
#como podemos observar tneemos dos columnas location esto se debe a que los nombres se supornen en ambos dataframes con distintos
#datos, en estos casos usamos un identificador unico que denonimos llave que puede ser compuesto por mas de una colummna para evitar
#este tipo de problemas
staff_df = pd.DataFrame([{'Name' : 'Kelly', 'Lastname' : 'Desjarding' ,'Role' : 'Director of HR', 'Location' : 'State Street'},
                         {'Name' : 'Sally', 'Lastname' : 'Brooks', 'Role' : 'Course of Liason', 'Location' : 'Washington Avenue'},
                         {'Name' : 'James', 'Lastname' : 'Wilde', 'Role' : 'Grader', 'Location' : 'Washington Avenue'}])

student_df = pd.DataFrame([{'Name' : 'James', 'Lastname' : 'Hammond', 'School' : 'Bussines', 'Location' : '1024 Billiard Avenue'},
                           {'Name' : 'Mike' , 'Lastname' : 'Smith', 'School' : 'Law', 'Location' : 'Fraternity House #22'},
                           {'Name' : 'Sally', 'Lastname' : 'Brooks' , 'School' : 'Enginneering', 'Location' : '512 Wilson Crescent'}])


pd.merge(staff_df, student_df, how = 'inner', on=['Name', 'Lastname'])
```
