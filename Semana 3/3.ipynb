GROUP BY

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('..\Semana 2\census.csv')
df.head()
```

```python
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
%%timeit -n 3

for state in df['STNAME'].unique():
    # Use boolean indexing directly and calculate the average
    avg = np.average(df[df['STNAME'] == state]['CENSUS2010POP'])
    
    print(f'Counties in the state {state} have an average population of {avg}')
```

```python
%%timeit -n 3

# Como podemos ver, esto tardÃ³ un poco. Ahora hagamos una demostraciÃ³n del uso de groupby.
for group, frame in df.groupby('STNAME'):
    # Donde group es el agrupamiento mediante estado y frame el resto de la informaciÃ³n.
    avg = np.average(frame['CENSUS2010POP'])
    
    print("State: " + group + ", Average Population: " + str(avg))
```

```python
#vemos una mejora significativa en el tiempo de rendimiento
#otra forma en q podemos usar el gropu by es organizando los datos mediante una funcion 
#aunque para eso es necesario definir la columna de interes como indice
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group, frame in df.groupby(set_batch_number):
    print('There are ' + str(len(frame)) + ' record of group ' + str(group) + ' for procesing ')
```

```python
#como podemos ver la funcion toma el indice como el valor sobre el cual estamos construyendo las condiciones
#ahora probemos con otro ejemplo
df = pd.read_csv('listings.csv')
df.head()
```

```python

print("id" in list(df.columns))
```

```python
#Otra cosa q se nos podria ocurrir es q sucede si tengo un indice multiple bueno en este caso habria q escribir
#cual consideramos como el agrupador mediante el parametro level
df=df.set_index(["cancellation_policy","review_scores_value"])

for group, frame in df.groupby(level=(0,1)):
    print(group)
```

```python
#tambien podemos aplicar funcion en estos dataframes de indices multiples
def grouping_fun(item):
    if item[1] == 10:
        return (item[0], "10.00")
    else:
        return (item[0], "not 10.00")
    
for group, frame in df.groupby(grouping_fun):
    print(group)
```

```python
#ahora veremos los 3 procesos que usualmente se usan en el procesamiento de datos, tales como agrupamiento, transformacion y filtracion de datos
```

AGREGACION


```python
df = df.reset_index()

#en este punto aprenderemos a usar la funcion agg que nos permite pasar como parametros un diccionario sobre los datos q nos interesa 
df.groupby("cancellation_policy").agg({"review_scores_value" : np.average})
```

```python
#dada la naturaleza de algunos de los datos en este caso la funcion average no es la mas optima entonces:
df.groupby("cancellation_policy").agg({"review_scores_value" : np.nanmean})
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value" : (np.nanmean,np.nanstd),
                                       "reviews_per_month" : np.nanmean})
```

TRANSFORMACION

```python
cols = ['cancellation_policy', "review_scores_value"]
#en este caso aplicamos el transform para hallar la media de cada grupo individual de cancellation policy, ademas este resultado
#se reemplaza en el df original
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
#dado q hallamos la media de review_scores_value cambiamos el nombre para hacerlo mas descriptivo, finalmente juntamos los dataframes
transform_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)
df=df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

```python
#y ahora calculamos la diferencia entre las puntaciones originales y la media q calculamos
df['mean_diff']=np.absolute(df['review_scores_value']-df['mean_review_scores'])
df['mean_diff'].head()
```

FILTRACION


```python
#una muestra d efiltrado mediante el uso de filter
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

Applying

```python
df = pd.read_csv('listings.csv')
df = df[['cancellation_policy', 'review_scores_value']]
df.head()
```

```python
#anteriormente para hallar la diferencia de la media respecto a los valores originales tuvimos q agrupar, filtrar y transformar los datos
#con apply podriamos haber hecho todo esto
df = df.reset_index()
def means_review_scores(group):
    avg = np.nanmean(group['review_scores_value'])
    group['mean_score_value'] = np.abs(avg - group['review_scores_value'])
    
    return group

df.groupby(['cancellation_policy']).apply(means_review_scores).head()
```
