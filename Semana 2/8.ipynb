Manipulating Dataframes

```python
import pandas as pd

df = pd.read_csv('presidents.csv')
df.head(10)
```

```python
#dado q los datos son de wikipedia tenemos algunos pies de pagina que pueden ser molestos entonces primero limpiamos estos datos
#usando un patron 
df['First'] = df['President']
df['First'] = df['First'].replace("[ ].*", "", regex=True)
df.head()
```

```python
#como podemos ver hemos logrado dividir el primer nombre del presidente en otra columna
#sin embargo hay otra forma de lograr lo mismo de una manera mas sencilla
del(df['First'])

#creamos la funcion o accion que deseamos aplicar a la columna objetivo
def splitname(row):
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    
    return row

df = df.apply(splitname, axis='columns')
df.head()
```

```python
#como podemos ver hemos obtenido el resultado esperado 
del(df['First'])
del(df['Last'])

#otra forma de lograr esta extracion de datos incluye nuevamente la construccion de patrones
pattern = '(^[\w]*)(?:.* )([\w]*$)'
df['President'].str.extract(pattern).head()
```

```python
#como podemos ver hemos obtenido la informacion que queriamos, sin embargo las columnas no tienen nombre podemos corregir esto
#de la siguiente manera
pattern = '(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)'
df['President'].str.extract(pattern).head()
```

```python
#ahora podemos extraer tambien los datos de la columnas born, en forma de mes, dia y año 
df['Born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})").head()

#df['Born']
```

```python
#y finalmente tenemos una funcion q nos permite cambiar el tipo a una fecha
df['Born'] = pd.to_datetime(df['Born'])
df['Born'].head()
```
