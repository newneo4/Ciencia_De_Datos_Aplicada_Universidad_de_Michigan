Indexing and Loading

```python
#En este tema aprenderemos a limpiar y procesar datos de un archivo csv
import pandas as pd

df = pd.read_csv('Admission_Predict.csv')
df.head()
```

```python
#como podemos ver pandas ha creado por defecto un indice para el dataframe
#si queremos definir una columna como indice para evitar necesitamos hacer lo siguiente
df = pd.read_csv('Admission_Predict.csv', index_col=0)#usamos la primera columna como indice
df.head()
```

```python
#y si queremos renombrar los nombre de las columnas podemos hacer lo siguiente
copy_df = df.rename(columns={'University Rating':'University Score'})
copy_df.head()
```

```python
#Tambien podemos listar los nombre de las columnas con el atributo columns
copy_df.columns
```

```python
#como podemos ver tenemos los nombre de todas las columnas y tambien observamos qye la columna
#LOR tiene un espacio incluido en el nombre y esto podria afectar en el momento de operar con esta
#entonces podmeos cambiarle de nommbre agregando dicho espacio a los parametros de nuestra funcion
#rename pero esto puede ser tedioso ademas, imaginen q tengamos 100 espacios exactos
```

```python
#entonces podemos abordar el problema limpiando el nombre 
copy_df = copy_df.rename(mapper=str.strip, axis='columns')
copy_df.columns
```

```python
#y como vemos hemos limpiado los espacios innecesarios en los nombres
```

```python
#Ahora si queremos cambiar los nombre de las columnas de manera conjunta a partir de los valores de una lista
#podemos seguir el siguiente procedimiento
cols = list(df.columns)
#en este caso eliminamos los espacios y las convertirmos en minusculas
cols = [x.lower().strip() for x in cols]

df.columns = cols
df.head()
```
