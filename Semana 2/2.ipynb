Consultas y ordenar series

```python
import pandas as pd
```

```python
#para consultar la localizacion mediante el indice numerico usamos
#iloc y para consultar mediante el label del indice usamos loc
students_classes = {'Alice': 'Physics',
                    'Jack' : 'Chemistry',
                    'Molly' : 'English',
                    'Sam' : 'History'}

s = pd.Series(students_classes)
s
```

```python
#iloc para el indice numerico de ubicacion
s.iloc[3]
```

```python
#loc para el indice label
s.loc['Molly']
```

```python
#Sin embargo python tiene una sintaxis inteligente y podemos pasar
#los parametros sin la necesidad de usar los metodos loc y iloc
#python reconocera q es lo q queremos hacer
s[3]
```

```python
s['Molly']
```

```python
#y para el manejo de operaciones podemos usar
grades = pd.Series([90,80,70,60])

total = 0

for grade in grades:
    total += grade
print(total/len(grades))
```

```python
#una manera mas eficiente es usando el soport te numpy
import numpy as np

total = np.sum(grades)

print(total / len(grades))
```

```python
#pero como podemos probar cual es mas rapida
numbers = pd.Series(np.random.randint(0,1000,10000))
numbers.head()
```

```python
len(numbers)
```

```python
%%timeit -n 100
total = 0
for number in numbers:
    total += number
    
print(total/len(numbers))
```

```python
%%timeit -n 100
total = np.sum(numbers)
print(total/len(numbers))
```

```python
#como podemos ver la diferencia de velocidad es bastante considerable
#y esto ocurre por q el segundo metodo uso un enfoque de vectorizacion
#lo q implica hacer varias tareas al mismo tiempo cargadas en la tarjeta
#grafica lo q reduce al final considerablemente el tiempo total
#esto tambien es considerado computacion paralela
```

```python
#podemos tener otro tipo de operaciones como aumentar 2 a los elementos d ela serie
numbers+=2
numbers.head()
```

```python
print("Antes de la actualización:", numbers)

for label, value in numbers.items():
    numbers[label] = value + 2

print("Después de la actualización:", numbers)
```

```python
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,1000))

for label, value in s.items():
    s.loc[label] = value+2
```

```python
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,1000))
s+=2
```

```python
#como podemos ver el tiempo es mucho menor y esto es dado que la mayoria
#de operaciones comunes estan vectorizadas todo esto implementado dentro de
#la biblioteca numpy
```
