Missing Values

Muchas veces tenemos valores que faltan dentro de un conjunto de datos, que puede tener justificacion de algun tipo y es necesario saber
como debemos manejar dichos valores


```python
import pandas as pd

df = pd.read_csv('class_grades.csv')
df.head(10)
```

```python
#una manera de filtrar estos datos es medinte el uso de la funcion isnull y usar sus resultados como mascara para el dataframe
mask = df.isnull()
mask
```

```python
#otra funcionque nos puede ayudar a filtrar filas con datos faltantes viene a ser dropna
df.dropna().head(10)
```

```python
#y si lo que queremos es llenar los valores na con un valor determinado podemos usar fillna
df.fillna(0, inplace=True)
df.head(10)
```

```python
#tambien muchas veces los datos filtrados como NaN puedes tener valores utiles que pandas interpreta como vacios
#una forma de corregir esto es la siguiente manera
df = pd.read_csv('log.csv')
df.head(10)
```

```python
df = df.set_index('time')
df = df.sort_index()
df.head(10)
```

```python
#como podemos observar hay usuarios que tienen el mismo indice entonces haremos un filtrado de indice multiple
df = df.reset_index()
df = df.set_index(['time', 'user'])
df.head(10)
```

```python
#una vez ordenado nuestro dataframe de manera adecuada podemos usar el llenado de valores NaN
df = df.fillna(method='ffill')
df.head()
```

```python
#y vemos que hemos llenado los valores faltante con los valores anteriores a estos gracias a la metodo ffill
df = pd.DataFrame({'A' : [1,1,2,3,4],
                   'B' : [3,6,3,8,9],
                   'C' : ['a','b','c','d','e']})

df
```

```python
#otra forma de reemplazar valores es mediante el metodo replace
df.replace(1,100)
```

```python
#tambien podemos cambiar valores mediante listas
df.replace([1,3],[100,300])
```
