Querying Dataframe

Una mascara de booleanos es un array que admite valores verdaderos o falsos y q usualmente es aplicado sobre otra estructura de datos basando sus salidas justamente en los valores que esta mascara tiena

```python
import pandas as pd

df = pd.read_csv('Admission_Predict.csv', index_col=0)
df.columns = [x.lower().strip() for x in list(df.columns)]

df.head()
```

```python
#creamos nuestra mascara de booleanos con la condicion q nos interesa
admit_mask = df['chance of admit'] > 0.7

admit_mask.head()
```

```python
#como podemos ver se ha filtrado los valores que cumplen con la condicion dada ahora podemos aplicar dicha mascara
#a nuestro dataframe con la funcion where
df.where(admit_mask).head()
```

```python
#como podemos observar los valores que no cumples con la condicion tiene valores NaN la manera de elimnar estos valores
#es mediante el uso de la funcion dropna
df.where(admit_mask).dropna().head()
```

```python
#tambien podemos tener una forma abreviada de hacer todos estoy pasos de la siguiente forma
df[df['chance of admit'] > 0.7].head()
```

```python
#la manera en q podemos pasar multiples mascara sobre un mismo dataframe es con el operador & para and
#en caso de que queramos que se cumplan necesariamente todas las condiciones dadas en las mascaras
# y | para or en caso de que queramos que solo sea necesario cumplir de nuestras condiciones para el filtro

(df['chance of admit'] > 0.7) & (df['gre score'] > 320)
```

```python
#es importante no olvidar el uso de parentesis en caso de que optemos por el uso del operador & o |
combined_mask = (df['chance of admit'] > 0.7) & (df['gre score'] > 320)
df[combined_mask]
```

```python
#otra sintaxis posible para resolver el mismo problema es la siguiente
altern_mash = df['chance of admit'].gt(0.7).lt(0.9)
```

```python
df[altern_mash]
```
